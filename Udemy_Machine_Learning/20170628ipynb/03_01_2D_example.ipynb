```python
# 正規分布に従う乱数の作り方。
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, # 1000個生成
                  n_features=2, # 2次元
                  centers=[(-5, -5), (0, 0), (5, 5)],    # クラスタ中心3個 で(-5,-5), (0,0), (5,5)がそれぞれの中心
                  cluster_std = 2, # クラスタの大きさ（標準偏差）
                  random_state=3   # 乱数種（再現用）
                 )
```

```python
# matplotlibの準備
import matplotlib.pyplot as plt
%matplotlib inline
plt.set_cmap(plt.cm.Paired) # 色設定
```

```python
# 2次元散布図でプロット
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k');
```

```python
# 境界線を引く関数の定義

def plotBoundary(X, clf, mesh=True, boundary=True, n_neighbors=1):

    # plot range
    x_min = min(X[:,0])
    x_max = max(X[:,0])
    y_min = min(X[:,1])
    y_max = max(X[:,1])

    # visualizing decision function
    #XX, YY = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200)) # make a grid
    XX, YY = np.mgrid[x_min:x_max:200j, y_min: y_max: 200j] # make a grid
    Z = clf.predict(np.c_[XX.ravel(), YY.ravel()]) # evaluate the value 
    
    Z = Z.reshape(XX.shape) # just reshape

    if mesh:
        plt.pcolormesh(XX, YY, Z, zorder=-10, alpha=0.4) # paint in 2 colors, if Z > 0 or not , zorder=-10 はグラフの重なり順序を表す。

    if boundary:
        plt.contour(XX, YY, Z, 
                    colors='k', linestyles='-', levels=[0]) # 等高線の本数と間隔は levels で指定できます。線を引きたい高さを配列で指定します
    print(Z)
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
```

```python
# モジュールの準備
from sklearn import neighbors

# オブジェクト作成
clf = neighbors.KNeighborsClassifier(n_neighbors=1)
```

```python
clf.fit(X, y) # 学習

plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')

plotBoundary(X, clf) # 境界線の描画
```

```python
# モジュールの準備
from sklearn import linear_model

# オブジェクト作成
clf = linear_model.LogisticRegression()
```

```python
clf.fit(X, y) # 学習

plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')

plotBoundary(X, clf) # 境界線の描画
```

```python
# モジュールの準備
from sklearn import svm

# オブジェクト作成
clf = svm.SVC(kernel='linear')
```

```python
clf.fit(X, y) # 学習

plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')

plotBoundary(X, clf) # 境界線の描画
```

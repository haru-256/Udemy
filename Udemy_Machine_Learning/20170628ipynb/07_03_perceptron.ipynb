````python
import numpy as np
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=20, # 20個生成
                  n_features=2, # 2次元
                  centers=2,    # クラスタ中心2個
                  cluster_std = 2, # クラスタの大きさ（標準偏差）
                  random_state=3   # 乱数種（再現用）
                 )
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.set_cmap(plt.cm.Paired);
````

````python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k');
````

````python
# 境界線を引く関数の定義

def plotBoundary(X, clf, mesh=True, boundary=True, n_neighbors=1):

    # plot range
    x_min = min(X[:,0])
    x_max = max(X[:,0])
    y_min = min(X[:,1])
    y_max = max(X[:,1])

    # visualizing decision function
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j] # make a grid
    
    Z = clf.predict(np.c_[XX.ravel(), YY.ravel()]) # evaluate the value 
    
    Z = Z.reshape(XX.shape) # just reshape

    if mesh:
        plt.pcolormesh(XX, YY, Z, zorder=-10) # paint in 2 colors, if Z > 0 or not

    if boundary:
        plt.contour(XX, YY, Z, 
                    colors='k', linestyles='-', levels=[0])
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
````

````python
from sklearn.linear_model import Perceptron

clf = Perceptron()
clf
````

````python
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')

plotBoundary(X, clf) # 境界線の描画
````

````python
X, y = make_blobs(n_samples=20, # 20個生成
                  n_features=2, # 2次元
                  centers=2,    # クラスタ中心2個
                  cluster_std =1, # クラスタの大きさ（標準偏差）
                  random_state=7   # 乱数種（再現用）
                 )
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k');

clf.warm_start = True # 以前のcoef_, intercept_ を使う。

for s in range(10):
    
    clf.random_state = 0 # データを見ていく順番を seed=0 で固定しておく
    
    clf.coef_, clf.intercept_ = np.random.rand(1,2) * 10 - 5, np.random.rand(1) * 10 + 30 # ここであらかじめ coef_, intercept_ を設定しておく。　
    # ランダムに初期値を変える。

    clf.fit(X,y) # warm_start = True なのであらかじめ設定しておいたcoef_, intercept_を　初期値として使用する。

    plotBoundary(X, clf, mesh=False)
    
plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')
````

毎回異なる初期値$\boldsymbol{w}, b$を設定するため得られる識別境界が異なる。

````python
X, y = make_blobs(n_samples=20, # 20個生成
                  n_features=2, # 2次元
                  centers=2,    # クラスタ中心2個
                  cluster_std =1, # クラスタの大きさ（標準偏差）
                  random_state=7   # 乱数種（再現用）
                 )
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k');

clf.warm_start = False

for s in range(10):

    clf.random_state = s
    
    clf.fit(X,y)

    plotBoundary(X, clf, mesh=False)
    
plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=y, edgecolors='k')
````

````python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target
````

````python
from sklearn.model_selection import StratifiedShuffleSplit
ss = StratifiedShuffleSplit(n_splits=1, 
                  train_size=0.8, 
                  test_size=0.2, 
                  random_state=0)

train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
print(np.unique(y_train, return_counts=True)[1] / y_train.size)
print(np.unique(y_test, return_counts=True)[1] / y_test.size)
````

````python
clf = Perceptron() # default max_iter=5, n_iter=None
clf.fit(X_train, y_train) # no scaling
clf.score(X_test, y_test)
````

````python
clf = Perceptron(warm_start=True, n_iter=1, random_state=0) # n_iter=1 を　max_iter  にしても良さそう。
# warm_start=True のため、前回のcoef_, intercept_　を初期値とする。
# そのため、一回のfitでn_iter(max_iter) エポックづつ学習していくことに相当する。

scores = []
n_range = range(1,50)
for n in n_range: # 1エポックづつ見ていく
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(n, score)
    scores.append(score)
scores = np.array(scores)
print(clf.coef_, clf.intercept_)
plt.plot(n_range, scores)
plt.xlabel("epochs")
````

for文を抜けた後に得られるスコアは
```python
clf.set_params(max_iter=49) # n_iter=49
```
とした時とだいたい同じもの
``coef_, intercept_ `` はデータ点を見ていく順番が異なるので、少し異なる。

````python
# その確認
clf.set_params(warm_start=False, max_iter=49, random_state=1).fit(X_train, y_train)
print('スコア', clf.score(X_test, y_test))
print(clf.coef_, clf.intercept_)
````

おそらく線形分離不可能なデータなのでエポックを増やしても識別率が収束しない

````python
# スケーリングを行ってみる。
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scale = scaler.transform(X_train)
X_test_scale  = scaler.transform(X_test)
````

````python
clf = Perceptron(warm_start=True, max_iter=1, random_state=0) # 比較のため、データを見ていく順番を固定する。

scores2 = []
n_range = range(1,50)
for n in n_range:
    clf.fit(X_train_scale, y_train)
    score = clf.score(X_test_scale, y_test)
    print(n, score)
    scores2.append(score)
scores2 = np.array(scores2)

plt.plot(n_range, scores, label="no scaling")
plt.plot(n_range, scores2, label="scaling")
plt.legend(loc="best")
plt.xlabel("epochs")
````

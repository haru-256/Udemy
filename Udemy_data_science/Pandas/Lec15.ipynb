```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
# NFLのデータをサンプルとして使います。
import webbrowser
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
# クリップボードから読み込むことが可能です。
nfl_frame1 = pd.read_clipboard()
nfl_frame1
```

```python
nfl_frame = DataFrame(nfl_frame1, columns=['Rank','Team','Won','Lost','Tied','Pct.','First','NFL','Season','Total','Games'])
nfl_frame
```

```python
# 列（カラム）の名前が.columnsでわかります。
nfl_frame.columns
```

```python
nfl_frame['Season']
```

```python
# オブジェクトの属性のような書き方も可能     空白がある場合には使えない
nfl_frame.Team
```

```python
nfl_frame[['Team','Season','Games']]
```

```python
# 特定のカラムで新しいDataFrameを作る
DataFrame(nfl_frame,columns=['Team','Season','Games']) #上の場合、参照なので変更すると元のDateFrameも変わってしまう。
```

```python
DataFrame(nfl_frame,columns=['Team','Season','Games','Stadium'])
```

```python
# indexを使って、行を取り出せる
nfl_frame.ix[3]
```

```python
# 列全体に値を代入
nfl_frame['Stadium']="Levi's Stadium" #文字列内に「'」があるので、全体は「”」で囲む Stasiumはnfl_frameにないので追加される。
```

```python
nfl_frame
```

```python
# 長さが合っていれば、列への代入が可能。
nfl_frame["Stadium"] = np.arange(6)
nfl_frame
```

```python
# SeriesをDataFrameに追加する
stadiums = Series(["Levi's Stadium","AT&T Stadium"],index=[4,0])
```

```python
stadiums
```

```python
nfl_frame['Stadium']=stadiums #indexに応じて値が入っている
nfl_frame
```

```python
# 列を消すことも可能
del nfl_frame['Stadium']
nfl_frame
```

```python
# 辞書からDataFramesを作ることもできます。
data = {'City':['SF','LA','NYC'],
        'Population':[837000,3880000,8400000]}

city_frame = DataFrame(data, columns=['City', 'Population'])  #dictionaryだと順番を保持しないのでcolumnsで指定すると良い。
#Python3.6からはdictionaryにも順番がある
#Show
city_frame
```

```python
import numpy as np
from numpy.random import randn

from scipy import stats

import seaborn as sns

%matplotlib inline
```

```python
# 英語のページですが・・・
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'

data1 = randn(100)
data2 = randn(100) + 2 # Off set the mean ＋２：平均を＋２する
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
# boxplot(箱ひげ図）を作ります。
sns.boxplot(data=[data1,data2])  # 引数データにdata=[x, y] とリストで渡すと２つとも描いてくれる。  点は外れ値
```

```python
# 外れ値を含むことができます。
sns.boxplot(data=[data1,data2],whis=np.inf)  # whis = np.inf とするとヒゲの中に外れ値も含む。
```

```python
# 水平に並べることもできます。 orient='h' (horizontal)水平　'v'(vertical) 垂直    orient: 方向
sns.boxplot(data=[data1,data2],whis=np.inf, orient = 'h')  
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf, orient = 'v')  
```

```python
# Violin/Viola plot はbox plotのシンプルさとKDEの情報量を兼ね備えます。
```

```python
# 正規分布に従う乱数です。  平均　０、標準偏差５の正規分布に従う乱数を１００個生成 rvs:ランダムにとるメソッド
data1 = stats.norm(0,5).rvs(100)

# γ分布に従う乱数を生成します。  np.concatenate(array_like=...)  array_like という型は、配列をndarrayの他、リストやタプルで表現したもの。
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# ヴァイオリンプロットを描くと違いが分かります。
sns.violinplot(data =[data1,data2])  # boxplotでは分布の違いがわからなかったが、violinplotでは違いがわかる。
```

```python
# violinplotはKDEによって作られている。　バンド幅を細かくしてみましょう。
sns.violinplot(data=data2,bw=0.01)
```

```python
# rugプロットと同じような機能もあります。 他にもinner='stick'、'box', 'quartile': 四分位, 'point'がある  defaultはbox
sns.violinplot(data=data1,inner="stick")
```

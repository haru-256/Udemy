```python
from numpy.random import randn
import seaborn as sns
import matplotlib
# Webブラウザに画像を埋め込むため
%matplotlib inline
```

```python
dataset = randn(100)
#matplotlib.rcParams["patch.force_edgecolor"] = True  # ヒストグラム
# seaborn.ditplot()はカーネル密度推定とヒストグラムを描きます   distribution(分布)
sns.distplot(dataset,bins=25)
```

```python
sns.distplot(dataset,rug=True,hist=False)  # histは描かずに、rugplotとKDEをかく
```

```python
# distplotは複数のグラフを書くので、それぞれのグラフのスタイルを変更することもできます。
sns.distplot(dataset,bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},   # kde_kws はKDEに対しての引数
             hist_kws={'color':'blue','label':"HISTOGRAM"})      # hist_kws はヒストグラムに対しての引数
```

```python
from pandas import Series

ser1 = Series(dataset,name='My_DATA')
```

```python
# pandasのSeriesを描画します
sns.distplot(ser1,bins=25)  # 軸名も描いてくれる。distplotはpandasではSeriesのみ引数に取れる。詳しくはドキュメント参照　　seaborn はpandasとも相性がいい。
```

```python
import numpy as np
from numpy.random import randn
import pandas as pd

# scipyの統計パッケージも使います。
from scipy import stats

# 描画のためのライブラリです。
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns  #　importするだけで、綺麗なグラフになる。

# ブラウザ内に画像を埋め込むための命令です
%matplotlib inline
sns.set_style('darkgrid')
```

ヒストグラムの説明は以下を参照してください。
https://ja.wikipedia.org/wiki/%E3%83%92%E3%82%B9%E3%83%88%E3%82%B0%E3%83%A9%E3%83%A0

```python
# 正規分布に従う乱数を作ります。
np.random.seed(1)
dataset1 = randn(100)
# bins=10がデフォルトです。
plt.hist(dataset1)   # ec(edgecolor)を指定しないと区切り線が透明になる。欲しい時はec='black'とすれば良い　　normed=Falseなので　縦軸は度数
```

```python
# もう一つ作ります。
dataset2 = randn(80)
# 色を変えてみましょう。  引数の詳しくはドキュメント参照
plt.hist(dataset2,color='indianred')
```

```python
plt.hist(dataset1, normed=True, ec='black')   # normed=True で正規化（積分した時に１になるように）する　　縦軸が変化
```

```python
# We can use normed to plot on same plot
# 正規化すると他のヒストグラムと重ねて書くことができる。
# Set normed=True for the plots to be normalized in order to comapre data sets with different number of observations
# Set alpha=0.5 for transperancy(透明度)

plt.hist(dataset1,normed=True,alpha=0.5,bins=20, ec='black')
plt.hist(dataset2,normed=True,alpha=0.5,bins=20, color='indianred', ec='black')
```

```python
import matplotlib.pyplot as plt
#plt.rcParams["patch.force_edgecolor"] = True  # ヒストグラムにおいて区切り線を表示させる。  上でec='black'引数は取らなくてもよくなる。
data1 = randn(1000)
data2 = randn(1000)

# 同時分布（結合分布） がsns.jointplot()メソッドでかける。
sns.jointplot(data1, data2, edgecolor='white')# ヒストグラムにおいてec(edgecolor)は使えない  edgecolorは散布図の区切り線を表示させるためのもの。
# x = np.linspace(-3, 3, 1000)  上に合わせて図をかけない？
```

```python
# もうすこし分かり易く
sns.jointplot(data1,data2,kind='hex')
```

```python
sns.jointplot(data1, data2, kind='kde')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-1, 1, 100)
y = x*x
ax.plot(x, y, label='二次関数')
ax.set_xlabel('x 軸')
ax.set_ylabel('y 軸')
plt.legend()
```
